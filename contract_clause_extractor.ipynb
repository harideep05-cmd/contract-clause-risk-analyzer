```python
def summarize_clause_with_severity(clause_text):
    prompt = f"""
You are assisting contract review.

Given the following legal clause:
1. Summarize the MAIN RISK or OBLIGATION in simple language.
2. Assign a qualitative RISK SEVERITY: Low, Medium, or High.
3. Provide a one-line justification for the severity.

Rules:
- Do NOT give legal advice.
- Do NOT mention laws or jurisdictions.
- Be concise.
- Output in the following format ONLY:

Risk Summary:
- bullet point 1
- bullet point 2
- bullet point 3

Risk Severity: <Low/Medium/High>

Reason: <one sentence>

Clause:
{clause_text}
"""
    response = model.generate_content(prompt)
    return response.text
```

```python
# Take one clause from the sample
test_clause = sample_df.iloc[0]['clause_text']
test_type = sample_df.iloc[0]['clause_type']

# Generate risk summary + severity
output = summarize_clause_with_severity(test_clause)

print("CLAUSE TYPE:")
print(test_type)

print("\nORIGINAL CLAUSE (partial):")
print(test_clause[:400], "...")

print("\nAI OUTPUT:")
print(output)
```

```python
new_clause = """
The Receiving Party shall not disclose any confidential information obtained under this Agreement
to any third party without prior written consent. This obligation shall remain in effect for three
years after termination of the Agreement.
"""
print(summarize_clause_with_severity(new_clause))
```

```python
new_clause = """
The Service Provider shall be liable for all direct and indirect damages arising from any failure
to perform its obligations under this Agreement, without limitation.
"""
print(summarize_clause_with_severity(new_clause))
```
